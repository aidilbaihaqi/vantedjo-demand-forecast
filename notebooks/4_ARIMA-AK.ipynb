# ARIMA Forecasting untuk Ayam Kampung

Notebook ini melakukan pemodelan ARIMA untuk time series **Ayam Kampung** dan melakukan forecast 14 hari ke depan.

```
# ===============================
# 1. Import Library
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

%matplotlib inline
```

## 2. Load & Persiapan Data

```
# Ganti path ini jika file berada di lokasi berbeda
file_path = "ts_ayam_kampung_clean.csv"

# Baca data
df = pd.read_csv(file_path)

# Konversi kolom date menjadi datetime
df['date'] = pd.to_datetime(df['date'])

# Jadikan kolom date sebagai index
df = df.set_index('date')

# Sort index (jaga-jaga jika belum urut)
df = df.sort_index()

print("Head data:")
print(df.head())
print("\nInfo:")
print(df.info())
```

## 3. Visualisasi Time Series Ayam Kampung

```
plt.figure(figsize=(12, 4))
plt.plot(df['Ayam_Kampung'])
plt.title("Time Series Penjualan/Stok Ayam Kampung")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Ayam Kampung")
plt.tight_layout()
plt.show()
```

## 4. Pembangunan Model ARIMA

```
# Ambil series target
series = df['Ayam_Kampung']

# Bangun model ARIMA(1,1,1)
model = ARIMA(series, order=(1, 1, 1))
model_fit = model.fit()

print("\nRingkasan Model ARIMA(1,1,1):")
print(model_fit.summary())
```

## 5. Forecast 14 Hari ke Depan

```
n_forecast = 14

# Forecast 14 hari ke depan
forecast_values = model_fit.forecast(steps=n_forecast)

# Buat index tanggal untuk 14 hari ke depan
last_date = df.index.max()
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1),
                             periods=n_forecast,
                             freq='D')

# Jadikan ke bentuk Series dengan index tanggal
forecast_series = pd.Series(forecast_values.values,
                            index=future_dates,
                            name="Forecast_Ayam_Kampung")

print("Forecast 14 hari ke depan:")
print(forecast_series)
```

## 6. Plot Historis + Forecast

```
plt.figure(figsize=(12, 4))
plt.plot(df['Ayam_Kampung'], label='Historis')
plt.plot(forecast_series, label='Forecast 14 Hari', linestyle='--')
plt.title("Historis vs Forecast Ayam Kampung (14 Hari ke Depan)")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Ayam Kampung")
plt.legend()
plt.tight_layout()
plt.show()
```

## 7. Simpan Hasil Forecast ke CSV

```
output_forecast = forecast_series.reset_index()
output_forecast.columns = ['date', 'Forecast_Ayam_Kampung']

output_path = "forecast_ayam_kampung_14hari.csv"
output_forecast.to_csv(output_path, index=False)
print(f"Forecast sudah disimpan ke: {output_path}")
```
